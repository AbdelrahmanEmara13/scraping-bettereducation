```
url='https://bettereducation.com.au/school/primary/nsw/nsw_top_primary_schools.aspx'
response= requests.get(url)
soup= BeautifulSoup(response.text, 'lxml')
print(soup.title.text)
 
 
 
 
schoolsTable = soup.find("table")
schools=schoolsTable.find_all('tr')[1:]
df = pd.DataFrame(columns=[ 
                'school' ,
                'State_overall_score',
                 'English' ,
                 'Maths',
                 'Totalenrolments',
                 'Trend/Compare',
                 'Locality' ,
                 'Sector'])
counter=0
clicks=0
dummylist=[]
for school in schools:
  schoolName=school.find_all('td')[0].text
  overallScore=school.find_all('td')[1].text
  English=''#school.find_all('td')[2].find('img').attrs['src'][-5]
  Maths=''#school.find_all('td')[3].find('img').attrs['src'][-5]
  enrolments=school.find_all('td')[4].text
  trend=school.find_all('td')[5].text
  Locality=school.find_all('td')[6].text
  Sector=school.find_all('td')[7].text
 
  driver.get('https://bettereducation.com.au/school/primary/nsw/nsw_top_primary_schools.aspx')
  if counter == 10:
    clicks= clicks+1
    counter=0
  print(clicks)
  for i in range(clicks):
  
      driver.find_element_by_partial_link_text('Next').click()
      
  url=str(schoolName).strip()
  driver.find_element_by_partial_link_text(url).click()
 
  print(driver.current_url)
  Durl= driver.current_url
  response= requests.get(Durl, allow_redirects=True)   
  soup= BeautifulSoup(response.text, 'lxml')
  try:  
    courseRange=soup.find("td", text='Course Range (Levels)').find_next_sibling("td").text
  except:
    courseRange='Empty'
    
  
  try:  
    coEducation=soup.find("td", text='Co-Education').find_next_sibling("td").text
  except:
    coEducation='Empty'
    
  
  try:  
    courseRange=soup.find("td", text='Course Range (Levels)').find_next_sibling("td").text
  except:
    courseRange='Empty'
  
  try:  
    denomination=soup.find("td", text='Denomination').find_next_sibling("td").text
  except:
    denomination='Empty'
 
  try:  
    students=soup.find("td", text='Students (approximate)').find_next_sibling("td").text
  except:
    students='Empty'
 
 
  df = df.append({'school': schoolName ,'State_overall_score':overallScore, 'English':English ,'Maths':Maths,'Totalenrolments':enrolments,'Trend/Compare':trend,'Locality':Locality ,'Sector':Sector, 'courseRange':courseRange, 'coEducation':coEducation, 'denomination': denomination, 'students':students}, ignore_index=True)
  print(len(df))
  counter=counter+1
 
  if len(df)==603:
    df.to_csv('Top Primary Schools in NSW - 2019.csv')
    !cp data.csv "drive/My Drive/"
```
